# Degree 2 

###  y = m1 (x1) + m2 (x^2) + c

```python
Independent_Variable = np.array([1,1.5,2,2.5,3,4,4.5,5,5.5,6,6.5,7,7.5,8])

Independent_Variable_sq=Independent_Variable**2

X = np.column_stack((Independent_Variable,Independent_Variable_sq))
y = Dependent_Variable 

regr.fit(X,y)

m1 = regr.coef_[0]
m2 = regr.coef_[1]
c = regr.intercept_

y_pred_2 = (m1*Independent_Variable)+(m2*Independent_Variable_sq)+c

print("Score:",round(r2_score(y,y_pred_2),4)*100)

plt.figure(figsize=(8,5))
plt.scatter(Independent_Variable, Dependent_Variable,  color='red')
plt.plot(Independent_Variable,y_pred_2)
plt.xlabel('Independent Variable/Input Parameter')
plt.ylabel('Dependent Variable/ Output Parameter')
plt.show()
```

# Degree 3

###  y = m1 (x1) + m2 (x^2) + m3  (x^3) + c
 

```python
Independent_Variable = np.array([1,1.5,2,2.5,3,4,4.5,5,5.5,6,6.5,7,7.5,8])
Dependent_Variable = np.array([-22,-10, -8,-5, 2, 8,7, 10,11 ,8,9, 2,0, -8])

Independent_Variable_sq=Independent_Variable**2
Independent_Variable_cube = Independent_Variable**3

X = np.column_stack((Independent_Variable,Independent_Variable_sq,Independent_Variable_cube))
y = Dependent_Variable 

regr.fit(X,y)

m1 = regr.coef_[0]
m2 = regr.coef_[1]
m3 = regr.coef_[2]

c = regr.intercept_

y_pred_3 = (m1*Independent_Variable)+(m2*Independent_Variable_sq)+(m3*Independent_Variable_cube)+c

print("Score:",round(r2_score(y_pred_3,y),4)*100)

plt.figure(figsize=(8,5))
plt.scatter(Independent_Variable, Dependent_Variable,  color='red')
plt.plot(Independent_Variable,y_pred_3)
plt.xlabel('Independent Variable/Input Parameter')
plt.ylabel('Dependent Variable/ Output Parameter')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error


X=np.array([1,1.5,2,2.5,3,4,4.5,5,5.5,6,6.5,7,7.5,8])
y=np.array([-22,-10, -8,-5, 2, 8,7, 10,11 ,8,9, 2,0, -8])

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)

# 1D to 2D
X_train=X_train[:,np.newaxis]
X_test = X_test[:,np.newaxis]


lr = LinearRegression()

lr.fit(X_train,y_train)

y_train_pred = lr.predict(X_train)
y_test_pred = lr.predict(X_test)


# print(np.sqrt(mean_squared_error(y_train,y_train_pred)))
train_error = np.sqrt(sum((y_train_pred - y_train)**2)/len(y_train))


# print(np.sqrt(mean_squared_error(y_test,y_test_pred)))
test_error = np.sqrt(sum((y_test - y_test_pred)**2)/len(y_test))

print("Train root mean sqr Error:",train_error)
print("Test root mean sqr Error:",test_error)

print("Train accuracy:",r2_score(y_train,y_train_pred)*100)
print("Test accuracy:",r2_score(y_test,y_test_pred)*100)

print("slop:",lr.coef_[0])
print("intercept:",lr.intercept_)

plt.figure(figsize=(8,5))
plt.scatter(X, y,  color='red')
plt.plot(X_train,X_train*lr.coef_[0]+lr.intercept_)
plt.xlabel('Independent Variable/Input Parameter')
plt.ylabel('Dependent Variable/ Output Parameter')
plt.show()
```
