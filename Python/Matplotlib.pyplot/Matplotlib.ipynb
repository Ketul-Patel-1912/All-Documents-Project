# Bar

```python
classes = ['pyhton' , 'R' , 'AI' , 'ML' , 'DS']

class1_students = [30,10,40,25,10]
class2_students = [40,30,50,15,50]
class3_students = [50,20,10,45,70]

#plt.barh (horinzontal)
plt.bar(classes,class1_students,width=0.3,align='center',color='y',
       edgecolor='m',linewidth=1.5,alpha=0.9,linestyle=':',
       label='class-1 Student')

plt.legend()
plt.show()
```

```python
import numpy as np

classes = ['pyhton' , 'R' , 'AI' , 'ML' , 'DS']

class1_students = [30,10,40,25,10]
class2_students = [40,30,50,15,50]
class3_students = [50,20,10,45,70]

w=0.2
bar1 = np.arange(len(classes))
bar2 = [i+w for i in bar1]
bar3 = [i+w for i in bar2]

plt.figure(figsize=(13,6),facecolor='gray')
#plt.barh (horinzontal)
plt.bar(bar1,class1_students,width=w,label='class-1 Student'
        ,align='center',color='b',edgecolor='m',linewidth=1.5,
        alpha=0.9,linestyle=':')

plt.bar(bar2,class2_students,width=w,label='class-2 Student',
        align='center',color='r',edgecolor='m',linewidth=1.5,
        alpha=0.9,linestyle=':')

plt.bar(bar3,class3_students,width=w,label='class-3 Student',
        align='center',color='g',edgecolor='m',linewidth=1.5,
        alpha=0.9,linestyle=':')


plt.xticks(bar2,classes,color='w',fontsize=15,fontweight='bold')
plt.yticks(color='w',fontsize=10,fontweight='bold')

plt.legend()
plt.show()
```

```python
import numpy as np

classes = ['pyhton' , 'R' , 'AI' , 'ML' , 'DS']

class1_students = [30,10,40,25,10]
class2_students = [40,30,50,15,50]
class3_students = [50,20,10,45,70]

w=0.2
bar1 = np.arange(len(classes))
bar2 = [i+w for i in bar1]
bar3 = [i+w for i in bar2]

plt.figure(figsize=(13,8),facecolor='gray')
#plt.barh (horinzontal)
plt.barh(bar1,class1_students,w,label='class-1 Student'
        ,align='center',color='b',edgecolor='m',linewidth=1.5,
        alpha=0.9,linestyle=':')

plt.barh(bar2,class2_students,w,label='class-2 Student',
        align='center',color='r',edgecolor='m',linewidth=1.5,
        alpha=0.9,linestyle=':')

plt.barh(bar3,class3_students,w,label='class-3 Student',
        align='center',color='g',edgecolor='m',linewidth=1.5,
        alpha=0.9,linestyle=':')


plt.yticks(bar2,classes,color='w',fontsize=15,fontweight='bold')
plt.xticks(color='w',fontsize=10,fontweight='bold')

plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = [[5., 25., 50., 20.],
  [4., 23., 51., 17.],
  [6., 22., 52., 19.]]

color_list = ['b', 'g', 'r']
gap = .8 / len(data)

for i, row in enumerate(data):
  X = np.arange(len(row))
  plt.bar(X + i * gap, row,
    width = gap,
    color = color_list[i % len(color_list)])

plt.show()
```

# pie chart

```python
import matplotlib.pyplot as plt

classes = ['pyhton' , 'R' , 'AI' , 'ML' , 'DS']

class1_students = [30,10,40,25,10]

explot = [0.04,0,0.06,0,0]

txt={'fontsize':18}
# plt.figure(figsize=(15,6))
plt.pie(class1_students,labels=classes,autopct="%0.0f%%",startangle=90,explode=explot,shadow=True,textprops=txt,radius=1.5)
plt.show()
```
