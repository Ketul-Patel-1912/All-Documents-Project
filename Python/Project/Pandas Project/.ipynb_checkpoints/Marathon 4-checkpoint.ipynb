```python
df.Votes=drop(sorted(df.Votes.unique())
```

```python
df['Votes']=df.Votes.map(lambda x:x.replace("-","10")).astype(int)
```

## Extract city name from Location column and store in a new column named city

```python
df.head()
```

```python
def location_split(location):
    print(location.spilt(",")[2])

df["City"]=df['Location'].map(lambda x:x.split(",")[-1])
```

```python
df.head()
```

## Find Average Average_cost in each City 

```python
# df.groupby('City')['Average_Cost'].mean()

df.groupby('City').agg({'Average_Cost':["mean","max","min"],"Minimum_Order":["max","min","mean","quantile"]})
```

## Find the Average Delivery_Time for Restaurants with Rating greater than 3.5 

```python
df.head(2)
```

```python
# df['Rating'].unique()

# - == 2.0
# new == 0.0
# temp close == 3.0
# Opening Soon == 2.1


df.loc[df.Rating=="-","Rating"]="2.0"
df.loc[df.Rating=="NEW","Rating"]="0.0"
df.loc[df.Rating=="Opening Soon","Rating"]="2.1"
df.loc[df.Rating=="Temporarily Closed","Rating"]="3.0"

df['Rating']=df['Rating'].astype(float)

df['Rating'].unique()
```
