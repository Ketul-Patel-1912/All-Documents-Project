## Outer Join (Concat)

 pd.concat([df1,df2])


```python
a = {'name':['hardik','ketul'],'age':[27,28]}
b= {'name':['ravi','rahul'],'age':[20,21]}

df1 = pd.DataFrame(a)
df2 = pd.DataFrame(b)

pd.concat([df1,df2])

```

## Inner Join using concat

pd.concat([df1,df2],axis=1,join='inner') 

```python
pd.concat([df1,df2],axis=0,join='inner')
```

```python
pd.concat([df1,df2],axis=1,join='inner')
```

## Inner Join using merge

pd.merge(df1,df2,on=['name'],how='inner')

```python
a = {'name':['hardik','ketul'],'age':[27,28]}
b= {'name':['hardik','ravi'],'age':[27,21]}

df1 = pd.DataFrame(a)
df2 = pd.DataFrame(b)


pd.merge(df1,df2,on=['name'],how='inner')
```

```python
pd.merge(df1,df2,on=['name'],how='left')
```

```python
pd.merge(df1,df2,on=['name'],how='right')
```

# DataSet Import

```python
import pandas as pd
cars=pd.read_csv('mtcars.csv')
cars.head()
```

```python
cars.shape
```

```python
cars.info(null_counts=True)
```

```python
cars.describe()
```

```python
cars.count() ## Not null record
```

## Data Cleaning

```python
## renaming the column name


# df.rename(columns={"OldName":"NewName"})
# df.rename(columns=lambda x: x.split("_")[1])

# df.set_axis(new_column_list, axis=1)

# df.columns = ["new_column_1","new_column_2"]

cars=cars.rename(columns={'Unnamed: 0':'model'})
cars
```

```python
## changing column datatype 

# cars.mpg=cars.mpg.astype(float)

# cars.info(null_counts=True)
```

```python
## we have null value in qsec column

cars.qsec=cars.qsec.fillna(cars.qsec.mean())

cars.count()
```

```python
## dropping unwanted columns 

# cars=cars.drop(columns=['sr.No'])
```

```python
## Find correlation matrix

df=cars[['model', 'mpg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am',
       'gear', 'carb']].corr()
df
```

## Indexing by Position

```python
# first 4 record and all column  (Index location)

cars.iloc[:4,:]
```

```python
# all record for 2 column

cars.iloc[:,:2]
```

```python
cars.loc[0:10,["model","mpg"]]
```

```python
cars["am"]=2
cars.head()
```

```python
f=lambda x: x*2

cars["am"]=cars["am"].apply(f)

cars.head()
```

```python
## filtering

cars[cars['vs']==0]
```

```python
# Sorting

cars.sort_values(by='model',ascending=False)
```

```python
# cars['cyl']>6 # result in True False

filter1=cars['cyl']>6
filter_new=cars[filter1]
filter_new
```

```python
# filter multiple condition

filter2=(cars['cyl']>6) & (cars['hp']>180)
filter_new=cars[filter2]
filter_new
```
